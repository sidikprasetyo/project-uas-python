```python
# Menampilkan data unik dalam kolom 'Publisher'
publisher = df1['Publisher'].unique()
print("Data unik dalam kolom 'Publisher':", publisher)
```

```python
# Menghitung jumlah manga series berdasarkan publisher
publisher_manga_series = df1.groupby('Publisher')['Manga series'].count()
print(publisher_manga_series)
```

```python
# Menampilkan data unik dalam kolom 'Demographic'
demographic = df1['Demographic'].unique()
print("Data unik dalam kolom 'Publisher':", demographic)
```

```python
# Menghitung jumlah author berdasarkan demographic
author_demographic = df1.groupby('Demographic')['Author(s)'].count()
print(author_demographic)
```
